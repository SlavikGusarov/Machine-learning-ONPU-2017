## Кластеризация набора данных Mass-Shootings-Dataset.csv

```python
import graphlab
```

```python
data = graphlab.SFrame("Mass-Shootings-Dataset.csv")
```

```python
data.head()
```

### 1. Рассчитать TF-IDF для столбца Summary.

```python
data['tf-idf'] = graphlab.text_analytics.tf_idf(data["Summary"])
data
```

### 2. Заменить в столбце Mental Health Issues значения на [0, 1, 2], где:
### a) 0 - No
### b) 1 - Yes
### c) 2 - Unclear/Unknown и другие значения

```python
def changer(s):
    a = int()
    if s == 'No':
        a = 0
    elif s == 'Yes':
        a = 1
    elif s == 'Unclear' or s == 'Unknown':
        a = 2
    return a

data['Mental Health Issues'] = data['Mental Health Issues'].apply(changer)
```

```python
data.head()
```

### 3. На основе примера обучить модель с использованием такого набора фич:
### [‘tf_idf’, ‘Total victims’, ’Mental Health Issues’] для
### - 3 кластеров
### - 5 кластеров

```python
model_3 = graphlab.kmeans.create(data,
                               features=['tf-idf', 'Total victims', 'Mental Health Issues'],
                               num_clusters=3)
```

```python
model_5 = graphlab.kmeans.create(data,
                               features=['tf-idf', 'Total victims', 'Mental Health Issues'],
                               num_clusters=5)
```
