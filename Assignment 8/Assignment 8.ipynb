## Поиск похожего изображения с использованием Deep Features.

```python
import graphlab
```

```python
image_train = graphlab.SFrame("image_train_data/")
image_test = graphlab.SFrame("image_test_data/")
```

```python
graphlab.canvas.set_target('ipynb')
image_train['image'].show()
```

### 1) Выбрать два тестовых объекта на основе своей фамилии (написанной на латинице):

### image_index1 = ASCII (decimal) 1й буквы фамилии + ASCII (decimal) 3й буквы фамилии
### image_index2 = ASCII (decimal) 2й буквы фамилии + ASCII (decimal) 4й буквы фамилии

#### Gusarov

#### ASCII (decimal) 1й буквы фамилии = 71
#### ASCII (decimal) 2й буквы фамилии = 117
#### ASCII (decimal) 3й буквы фамилии = 115
#### ASCII (decimal) 4й буквы фамилии = 97

#### image_index1 = 71 + 115 = 186
#### image_index2 = 117 +  97 = 214

```python
image_index1 = 186
image_index2 = 214

image_one = image_test[image_index1:image_index1 + 1]
image_one.show()

image_two = image_test[image_index2:image_index2 + 1]
image_two.show()
```

### 2) Используя deep_features в image_train наборе данных, обучить модель nearest neighbor.

```python
model = graphlab.nearest_neighbors.create(image_train,
                                          features=['deep_features'],
                                          label='id')
```

### 3) Для каждого из двух тестовых объектов из пункта 1 подсчитать ближайшего соседа, используя метод graphlab.nearest_neighbors.NearestNeighborsModel.query построенной модели.

```python
neighbor_for_one = model.query(image_one, 'id', k=1)
neighbor_for_one
```

```python
neighbor_for_two = model.query(image_two, 'id', k=1)
neighbor_for_two
```

### 4) Отобразить полученные изображения.

```python
def get_images_from_ids(query_result):
    return image_train.filter_by(query_result['reference_label'],'id')

get_images_from_ids(neighbor_for_one).show()
get_images_from_ids(neighbor_for_two).show()
```
