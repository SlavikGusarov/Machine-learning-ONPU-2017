## Работа с NumPy
### Генерировать два случайных одномерных массива X, Y  длиной 30 элементов со случайными значениями в диапазоне [-20, 20]

```python
import numpy as np

x = np.random.randint(-20, 20 + 1, size=30)
y = np.random.randint(-20, 20 + 1, size=30)
print x
print y
```

## Сортировка
### Сортируем значения массива методом пузырек

```python
def buble_cort(array):
    swap_counter = -1
    while(swap_counter is not 0):
        swap_counter = 0
        for i in range(array.shape[0]):
            if array[i] > array[i+1]:
                array[i], array[i+1] = array[i+1], array[i]
                swap_counter +=1
            if array[i+1] == array[-1]:
                break
    return array

x = buble_cort(x)
print x
```

### Отрисуем полученные координаты в виде точек красного цвета

```python
plt.plot(x, y, 'ro')
plt.show()
```
